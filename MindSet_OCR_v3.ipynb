### Распознование ФИО с паспорта

Алгоритм построен на использование opencv и двух библиотек распознавания текста для сравнения tesseract и easyocr. Загружаем фотографии, вырезаем интересующую область, предварительно обработанную cv2.selectROI затем распознаем эту область

#### Основные библиотеки и функции

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

import easyocr
import pytesseract
```

```python
# Путь для подключения tesseract
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
def get_imlist(path,extension):
    """
    Возвращает список имен всех файлов в каталоге с заданным расширением
    """
    return [os.path.join(f)for f in os.listdir(path) if f.endswith(extension)]
```

```python
def plot_image(image):
    """
    Вывод изображения 
    """
#     plt.figure(figsize = (10,15))
    plt.imshow(image,cmap='gray')
    plt.show()
```

```python
def teseract_recognition(img):
    """
    Библиотека для оптического распознования текста
    """
    return pytesseract.image_to_string(img, lang='rus', config=r'--oem 3 --psm 6')
```

```python
def easyocr_recognition(img):
    """
    Библиотека для оптического распознования текста
    """
    return easyocr.Reader(["ru"]).readtext(img, detail=0, paragraph=False, text_threshold=0.8)
```

```python
my_dir='DataForOCR'
filelist=get_imlist(my_dir,'')
```

```python
# f_list=[]
# f_list=[my_dir+'/'+item for item in filelist]
```

#### Вырезаем область

```python
for image_list in filelist:
#     outfile=os.path.splitext(image_list)[0]
    #Читаем image
    image = cv2.imread(my_dir+'/'+image_list)
    # Предварительная обработка изображения переведенное в серый, размыто и преведено в негатив
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (3, 3), 0)
    rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (47, 47))
    blackhat = cv2.morphologyEx(blur, cv2.MORPH_BLACKHAT, rect_kernel)
    # Организуем окно вывода изображения, чтобы выбрать оптимально размер
    cv2.namedWindow("image", flags= cv2.WINDOW_NORMAL | cv2.WINDOW_FREERATIO)
    cv2.imshow("image", blackhat)
    # Выделяем ROI
    r = cv2.selectROI("image", blackhat)
 
    # Вырезаем image
    cropped_image = gray[int(r[1]):int(r[1]+r[3]),
                      int(r[0]):int(r[0]+r[2])]
 
    # Выводим на экран вырезанное изображение
    cv2.imshow("Cropped image", cropped_image)
    cv2.imwrite('Cropped/'+image_list,cropped_image)
    cv2.waitKey(0)
```

```python
filelist_cropped=get_imlist('Cropped','')
```

#### Распознаем область

```python
for image_list in filelist_cropped:
    img=cv2.imread('Cropped/'+ image_list)
    image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plot_image(image_gray)
    print(teseract_recognition(image_gray))
```

```python
# img=cv2.imread("Cropped image.jpg")
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

for image_list in filelist_cropped:
    img=cv2.imread('Cropped/'+ image_list)
    image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plot_image(image_gray)
    print(easyocr_recognition(image_gray))
```
